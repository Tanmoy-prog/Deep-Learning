```python
import keras
```

```python
from keras.preprocessing.image import ImageDataGenerator,img_to_array,load_img
```

```python
datagen=ImageDataGenerator(rotation_range=40,
                           width_shift_range=0.2,
                           height_shift_range=0.2,
                           shear_range=0.2,
                           zoom_range=.2,
                           horizontal_flip=True,
                           fill_mode='nearest'
                      )
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
img = load_img('/content/drive/MyDrive/file/cat.jpg',target_size=(224,224))
```

```python
img
```

```python
x = img_to_array(img)
```

```python
x= x.reshape((1,)+x.shape)
```

```python
x
```

```python
i=0
for a in datagen.flow(x,batch_size=1,save_to_dir='/content/drive/MyDrive/file',save_prefix='cat',save_format='jpeg'):
  i=i+1
  if i>=20:
    break
```
